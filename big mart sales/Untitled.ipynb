```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
train=pd.read_csv('Train.csv')
train.head()
```

```python
train.info()
```

```python
test=pd.read_csv('Test.csv')
test.info()
```

```python
train['Item_Weight']=train['Item_Weight'].fillna(train['Item_Weight'].median())
test['Item_Weight']=test['Item_Weight'].fillna(test['Item_Weight'].median())
train['Outlet_Size']=train['Outlet_Size'].fillna('Medium')
test['Outlet_Size']=test['Outlet_Size'].fillna('Medium')

```

```python
train.info()
```

```python
for col in 'Item_Fat_Content','Item_Type','Outlet_Size','Outlet_Location_Type','Outlet_Type':
    l=LabelEncoder()
    train[col]=l.fit_transform(train[col])
    test[col]=l.transform(test[col])
train.head()        
```

```python
l_item=LabelEncoder()
train['Item_Identifier']=l.fit_transform(train['Item_Identifier'])
test['Item_Identifier']=l.transform(test['Item_Identifier'])

l_outlet=LabelEncoder()
train['Outlet_Identifier']=l.fit_transform(train['Outlet_Identifier'])
test['Outlet_Identifier']=l.transform(test['Outlet_Identifier'])

```

```python
for col in 'Item_Fat_Content','Item_Type','Outlet_Size','Outlet_Location_Type','Outlet_Type':
    print(train.groupby(col).size())
    sns.violinplot(train[col],train['Item_Outlet_Sales'])
    plt.show()
```

```python
dat=train.loc[train['Item_Identifier']==100]
print(dat.groupby('Outlet_Size').size())
sns.violinplot(dat['Outlet_Size'],dat['Item_Outlet_Sales'])
plt.show()

rem=dat.loc[dat['Outlet_Size']==1].drop(['Outlet_Size','Item_Identifier'],axis=1)
for col in rem.columns:
    sns.violinplot(rem[col],rem['Item_Outlet_Sales'])
    plt.show()
```

```python
for col in test.columns: 
    if col not in ('Item_Fat_Content','Item_Type','Outlet_Size','Outlet_Location_Type','Outlet_Type'):
        sns.kdeplot(train[col],train['Item_Outlet_Sales'])
        plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(np.array(train.drop(['Item_Outlet_Sales'],axis=1),dtype='int')
                                               ,np.array(train['Item_Outlet_Sales'],dtype='int'))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
extra=RandomForestClassifier(n_estimators=200).fit(x_train,y_train).predict(x_test)
accuracy_score(y_test,extra)
```
